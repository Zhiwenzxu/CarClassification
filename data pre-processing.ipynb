```python
if not os.path.exists('png-files-kou2'):
    os.mkdir('png-files-kou2')
    
def save_data_png(fnames,bboxes):
    
    num_samples = len(fnames)
    src_folder = 'png-files-kou2'
    
    for i in range(15184,18185):
        (x1, y1, x2, y2) = bboxes[i]
        image = Image.open(os.path.join('png-files-15185', fnames[i]))
        print('saving ', fnames[i])
        imag = np.array(image)
        imag = imag[y1:y2,x1:x2]
        image = Image.fromarray(imag)
        imag = np.array(image.resize((100,100)))
        path = os.path.join(src_folder, fnames[i])
        Image.fromarray(imag).save(path)
```

```python
def split(index,df_train):
    df = df_train[df_train['class_label'] == index]
    df_shuffled = df.sample(n=df.shape[0], random_state=123)

    train_len = int(len(df_shuffled)*0.8)
    valid_len = int(len(df_shuffled)*0.15)

    df_trainset = df_shuffled.iloc[:train_len]
    df_validset = df_shuffled.iloc[train_len:train_len+valid_len]
    df_testset = df_shuffled.iloc[train_len+valid_len:]
    
    return df_trainset,df_validset,df_testset
```

```python
data_train = pd.DataFrame()
data_valid = pd.DataFrame()
data_test = pd.DataFrame()

for i in range(0,196):
    train,valid,test = split(i,df_train)
    data_train = data_train.append(train)
    data_valid = data_valid.append(valid)
    data_test = data_test.append(test)
```

```python
data_valid = data_valid.sample(n=data_valid.shape[0], random_state=123)
data_train = data_train.sample(n=data_train.shape[0], random_state=123)
data_test = data_test.sample(n=data_test.shape[0], random_state=123)
```

```python
data_train.to_csv('train1.csv', index=None)
data_valid.to_csv('valid1.csv', index=None)
data_test.to_csv('test1.csv', index=None)
```
